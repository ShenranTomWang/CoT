# Import

```python
from transformers import AutoModelForCausalLM
import torch
from utils.loading_utils import load_model_transformers, load_tokenizer
from utils.data_collection_utils import generate_single_line
import json
```

# Constants

```python
MODEL_NAME = "hymba-1.5b-instruct"
DATA_PATH = "./data/UMWP.jsonl"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

# Load Dataset

```python
with open(DATA_PATH, "r") as file:
    data = []
    for line in file:
        data.append(json.loads(line))
```

# Load Model

```python
model = load_model_transformers(MODEL_NAME, device, trust_remote_code=True)
tokenizer = load_tokenizer(MODEL_NAME, device)
output = generate_single_line(model, tokenizer, data[0]["question"], device=device)
```
